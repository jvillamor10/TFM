# Clustering

## KMEANS

Primeramente, se va aplicar el algoritmo kmeans

Se inicializan el parámetro de inicio y de iteración

```python
init = 'random' # initialization method 
iterations = 10
```

Se procede a calcular la distorsión y el silhouette para cada k-ésimo valor

```python
distortions = []
silhouettes = []

for i in range(2, 11):
    km = KMeans(n_clusters = i, init = init, n_init = iterations)
    labels = km.fit_predict(hr_window)
    distortions.append(km.inertia_)
    silhouettes.append(metrics.silhouette_score(hr_window, labels))
```

```python
x = [i for i in range(2,11)]
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('k')
ax1.set_ylabel('sse', color=color)
ax1.plot(x, distortions, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx() 

color = 'tab:blue'
ax2.set_ylabel('silhouette', color=color) 
ax2.plot(x, silhouettes, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()
```

Como se puede ver, el valor con mayor silhouette es k=3, pero, pensando en el problema y viendo que la diferencia entre el silhouette y el sse es muy parecido con K igual a 4, se escoge este último valor.

```python
k = 4
```

```python
print ("Número de clsuters", k)
km = KMeans(n_clusters = k, init = init)

y_km = km.fit_predict(hr_window)
```

Se guarda el modelo para usar posteriormente:

```python
file = open('../../modelos/clustering/hr/kmeans_model.pickle', 'wb')
pickle.dump(km, file)
file.close()
```

Ya se tienen las etiquetas

```python
km.labels_
```

Se asignan las etiquetas a los registros

```python
hr_window['cluster'] = km.labels_
```

```python
hr_window
```

Se agrupan los clusters y se hace una breve descripción para poder caracterizar los grupos y poder interpretar los datos.

```python
hr_window.groupby('cluster').describe()
```

Podemos ver que hay 4 grupos bien diferenciados, que están escalados de menor a mayor intensidad en la actividad.

Se van a renombrar los grupos para que queden ordenados de menor a mayor y así sea más fácil su interpretación.

```python
mapping = {0:2, 1:4, 2:3, 3:1}
```

```python
hr_window['cluster'] = [mapping[i] for i in hr_window['cluster']]
```

## Gaussian Mixture

Se procede a realizar clustering con el algoritmo Gaussian Mixture. Para ello, se prueba con distintos números de componentes y distintos tipos de covarianza, y nos quedamos con el que menor BIC tenga.

```python
gmm = GaussianMixture()
```

```python
fig, ax = plt.subplots(figsize=(6, 3.84))

n_components = range(1, 10)
covariance_types = ['spherical', 'tied', 'diag', 'full']

for covariance_type in covariance_types:
    valores_bic = []
    
    for i in n_components:
        modelo = GaussianMixture(n_components=i, covariance_type=covariance_type)
        modelo = modelo.fit(hr_window)
        valores_bic.append(modelo.bic(hr_window))
        
    ax.plot(n_components, valores_bic, label=covariance_type)
        
ax.set_title("Valores BIC")
ax.set_xlabel("Número componentes")
ax.legend();

```

El mejor valor es con el tipo de covarianza full y el número de componetes igual a 4, mismo número de componentes que en el caso anterior.

Se procede a obtener los clusters

```python
gmm = GaussianMixture(n_components=4, covariance_type='full')
y_gmm = gmm.fit_predict(hr_window)
```

```python
y_gmm
```

```python
hr_window['cluster'] = y_gmm
```

```python
hr_window.groupby('cluster').describe()
```

A priori, parecen peores resultados, un grupo demasiado pequeño y grupos muy homógeneos, con kmeans se ha podido diferenciar mejor lso grupos.

```python
hr_window.to_csv('../../datasets/hr-study/windows_10_50_k_4_gm.csv')
```

## Jerárquico

```python
dist = sklearn.neighbors.DistanceMetric.get_metric('euclidean')
mat_sim = dist.pairwise(hr_window)

ax = sns.heatmap(mat_sim,vmin=0, vmax=1)
```

```python
clusters = cluster.hierarchy.linkage(mat_sim, method = 'complete')
dend = cluster.hierarchy.dendrogram(clusters, labels = hr_window.index, color_threshold=7)

f = plt.figure()
plt.show()
```

```python
cut  = 10000
```

```python
clusters = cluster.hierarchy.linkage(mat_sim, method = 'complete')
dend = cluster.hierarchy.dendrogram(clusters, color_threshold=cut)
plt.figure(figsize = (20,15))
plt.show()
```

```python
labels = cluster.hierarchy.fcluster(clusters, cut , criterion = 'distance')
```

```python
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print(f'Número de clusters {n_clusters}')
print("Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(hr_window, labels))
```

Resultados:

```python
hr_window['cluster'] = labels
```

## DBSCAN

```python
hr_window.drop('cluster',axis=1,inplace=True)
```

En este apartado, se van a detectar los posibles outliers, para poder posteriormente estudiar comportamientos anómalos o bien por la jugador o bien por el sensor.

```python
dist = sklearn.neighbors.DistanceMetric.get_metric('euclidean')
mat_sim = dist.pairwise(hr_window)
```

```python
min_pts=100

A = kneighbors_graph(hr_window, min_pts, include_self=False)
Ar = A.toarray()

seq = []
for i,s in enumerate(hr_window):
    for j in range(len(hr_window)):
        if Ar[i][j] != 0:
            seq.append(mat_sim[i][j])
            
seq.sort()
```

```python
fig = px.line(x=numpy.arange(0, len(seq), 1), y=seq)
fig.show()
```
