Para cada tramo, se calculará el número de eventos realizados, solo el scroing puede ser negativo

```python
def retrieve_clean_section_attributes():
    return {
    'duration': 0,
    'foul_receives': 0,
    'fouls': 0,
    'lost_balls': 0,
    'goals': 0,
    'assists': 0,
    'shots_failed':0,
    '7m_suspensions_provoke':0,
    '7m_suspensions':0,
    '2_min':0,
    'scoring':0
}
```

```python
section_attributes = retrieve_clean_section_attributes()
section_attributes
```

Se define un método para obtener los eventos

```python
def retrieve_section_events(section_attributes, event, assist, offense_player, defense_player, scoring_play):
    if event == ' FOUL_RECEIVE' and offense_player == player_name:
        section_attributes['foul_receives'] += 1
    if event == ' FOUL_RECEIVE' and defense_player == player_name:
        section_attributes['fouls'] += 1
    if event == ' LOST_BALL' and offense_player == player_name:
        section_attributes['lost_balls'] += 1
    if assist == player_name:
        section_attributes['assists'] += 1
    if event == ' POST_OUT' or event == ' SAVE' and offense_player == player_name:
        section_attributes['shots_failed'] += 1
    if event == ' GOAL' and offense_player == player_name:
        section_attributes['goals'] += 1
    if event == ' 7M_PROVOKE' and offense_player == player_name:
        section_attributes['7m_suspensions_provoke'] += 1
    if event == ' 7M_PROVOKE_SUSPENSION' and offense_player == player_name:
        section_attributes['7m_suspensions_provoke'] += 1
    if event == ' 7M_PROVOKE_SUSPENSION' and defense_player == player_name:
        section_attributes['7m_suspensions'] += 1
    if event == ' 7M_PROVOKE' and defense_player == player_name:
        section_attributes['7m_suspensions'] += 1
    if event == ' 2_MIN' and offense_player == player_name:
        section_attributes['2_min'] += 1
    
    section_attributes['scoring'] += scoring_play
```

```python
for section in sections:
    arrojeria_events[(arrojeria_events['seconds'] >= section[0]) & (arrojeria_events['seconds'] <= section[1])]
```

```python
arrojeria_events[(arrojeria_events['seconds'] >= 667) & (arrojeria_events['seconds'] <= 1214)]
```

```python
for section in sections:
    section_attributes = retrieve_clean_section_attributes()
    section_attributes['duration']=section[1]-section[0]
    for index, row in arrojeria_events[(arrojeria_events['seconds'] >= section[0]) & (arrojeria_events['seconds'] <= section[1])].iterrows():

        retrieve_section_events(section_attributes, row['event'], row['assist'], row['offense_player'], row['defense_player'], row['impact_play'])
    new_df = new_df.append(section_attributes,ignore_index=True)
```

Como se puede ver en el dataset, se tiene por cada registro el tramo de juego, con la duración en segundos y cada evento que realiza, junto al scoring final del tramo.

Ahora hay que obtener la información de los clusters, para este caso se va a calcular el porcentaje de tiempo que se encuentra en cada grupo durante ese sector de juego.

```python
hr_window = pd.read_csv('../../datasets/hr-study/windows_10_50_k_4.csv',index_col=0)
```

```python
hr_window.head()
```

```python
hr_window.iloc[0]
```

```python
def retrieve_clean_section_attributes_clustering():
    return {
    'hr_low_percentage': 0,     
    'hr_medium_percentage': 0,
    'hr_high_percentage': 0,
    'hr_max_percentage': 0
}
```

```python
window = 10 * 50/100
```

```python
window
```

```python
clusters_dict = {
    1: 'low',
    2: 'medium',
    3: 'high',
    4: 'max'
}
```

```python
for key in retrieve_clean_section_attributes_clustering().keys():
    new_df[key] = ''
```

Se crean las columnas de las etiquetas de los clusters

```python
new_df
```

Se define el inicio del partido

```python
match_start = 3262
```

```python
section = sections[0]
```

```python
clusters = hr_window.iloc[round((section[0] + match_start)/2.5):round((section[1]+match_start)/2.5)].groupby('cluster').describe()['0']['count']
```

```python
clusters
```

```python
count = 0
cluster_map = {
    1: 0,
    2: 0,
    3: 0,
    4: 0
}
for index,value in clusters.iteritems():
    count += value
    cluster_map[index] = value
    print(index,value)
```

```python
for index in cluster_map:
    print(round(cluster_map[index]/count * 100, 2))
```

Lo que se va haciendo es crear un diccionario con los cuatro valores de las etiquetas, y se va sumando incrementando en uno el valor de la etiqueta por cada registro del conjunto de datos que se encuentre en el windowing. Mientras, se va guardando el número total de clusters analizados y se divide el número de clusters en ese registro por el número total, calculando así el porcentaje.

```python
count = 0
for section in sections:
    section_attributes = retrieve_clean_section_attributes_clustering()
    clusters = hr_window.iloc[round((section[0] + match_start)/2.5):round((section[1]+match_start)/2.5)].groupby('cluster').describe()['0']['count']
    clusters_count = 0
    cluster_map = {
        1: 0,
        2: 0,
        3: 0,
        4: 0
    }
    for index,value in clusters.iteritems():
        clusters_count += value
        cluster_map[index] = value
        
    for index in cluster_map:
        label = clusters_dict[index]
        section_attributes[f'hr_{label}_percentage'] = round(cluster_map[index]/clusters_count * 100, 2)
        
    new_df.at[count, 'hr_low_percentage'] = section_attributes['hr_low_percentage']
    new_df.at[count, 'hr_medium_percentage'] = section_attributes['hr_medium_percentage']
    new_df.at[count, 'hr_high_percentage'] = section_attributes['hr_high_percentage']
    new_df.at[count, 'hr_max_percentage'] = section_attributes['hr_max_percentage']
    count+=1
```

Se repite lo mismo pero con la actividad

```python
activity_load = pd.read_csv('../../datasets/activity-study/windows_5_60_al3d_k_4.csv',index_col=0)
```

```python
activity_load
```

```python
def retrieve_clean_section_attributes_clustering_al():
    return {
    'al_low_percentage': 0,     
    'al_medium_percentage': 0,
    'al_high_percentage': 0,
    'al_max_percentage': 0,
}
```

```python
for key in retrieve_clean_section_attributes_clustering_al().keys():
    new_df[key] = ''
```

```python
count = 0
for section in sections:
    section_attributes = retrieve_clean_section_attributes_clustering()
    clusters = activity_load.iloc[round((section[0] + match_start)/2.5):round((section[1]+match_start)/2.5)].groupby('cluster').describe()['0']['count']
    clusters_count = 0
    cluster_map = {
        1: 0,
        2: 0,
        3: 0,
        4: 0
    }
    for index,value in clusters.iteritems():
        clusters_count += value
        cluster_map[index] = value
        
    for index in cluster_map:
        label = clusters_dict[index]
        section_attributes[f'al_{label}_percentage'] = round(cluster_map[index]/clusters_count * 100, 2)
        
    new_df.at[count, 'al_low_percentage'] = section_attributes['al_low_percentage']
    new_df.at[count, 'al_medium_percentage'] = section_attributes['al_medium_percentage']
    new_df.at[count, 'al_high_percentage'] = section_attributes['al_high_percentage']
    new_df.at[count, 'al_max_percentage'] = section_attributes['al_max_percentage']
    count+=1
```
