# Cuaderno en el que se estudian los eventos que pueden ocurrir en el partido y como poder etiquetar el impacto de ese evento en el partido

```python
import pandas as pd
```

```python
pd.options.display.max_columns = None
```

```python
PLAY_BY_PLAY_DATASET_FILE_DIR = '../../datasets/original-data/Copia de playByplay-2023-05-10T20_05_08.442+02_00.csv'
```

```python
play_by_play_dataset = pd.read_csv(PLAY_BY_PLAY_DATASET_FILE_DIR, sep='\t')
play_by_play_dataset
```

```python
df = play_by_play_dataset[play_by_play_dataset[' Offense Player Name'] != ' '][[' Start', ' End', ' Event', ' Offense Player Name', ' Defense Player Name']]
df
```

Como primera idea, se pueden etiquetar los eventos en tres clases:
* -1: acción negativa
* 0: acción neutra
* 1: acción positiva

```python
df[' Event'].unique()
```

```python
df[df[' Event'] == ' FOUL_RECEIVE'].head()
```

Se puede interpretar que provocar una falta, para una jugadora ofensiva, es una acción positiva (etiqueta 1).

Para una jugadora defensiva, cometer una falta es algo más difuso, por tanto, se etiqueta como jugada neutra (etiqueta 0).

```python
df[df[' Event'] == ' POST_OUT'].head()
```

Para la jugadora ofensiva, lanzar al palo puede considerarse como una acción neutra (etiqueta 0).

Para la jugadora defensiva (portera), sería una jugada neutra (etiqueta 0).

```python
df[df[' Event'] == ' LOST_BALL'].head()
```

Para la jugadora ofensiva, perder un balón es una acción negativa (etiqueta -1).

No se registran las jugadoras defensivas en este evento.

```python
df[df[' Event'] == ' GOAL'].head()
```

Para una jugadora ofensiva, un gol es una jugada positiva (etiqueta 1).

Para una jugadora defensiva (portera), es una jugada negativa o neutra.

```python
df[df[' Event'] == ' 7M_PROVOKE'].head()
```

Para una jugadora ofensiva, provocar una falta de 7 metros es una jugada positiva.
Para una jugadora defensiva, provocar una falta de 7 metros es una jugada negativa.

```python
df[df[' Event'] == ' SAVE'].head()
```

Para una jugadora ofensiva, una parada es una jugada negativa (-1)

Para una jugadora defensiva (portera), una para es una jugada positiva (1).

```python
df[df[' Event'] == ' 7M_PROVOKE_SUSPENSION'].head()
```

Para una jugadora ofensiva, esto es una jugada positiva (1).

Para una jugadora negativa, esto es una jugada negativa (-1).

```python
df[df[' Event'] == ' 2_MIN']
```

En este caso, la jugadora ofensiva es la que provoca la exclusión de 2 minutos
