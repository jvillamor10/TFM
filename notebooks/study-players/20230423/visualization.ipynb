```python
def show_graphic(dataset, show_events, condition, number):
    # Colores para cada número
    colors = {1: 'blue', 2: 'green', 3: 'yellow', 4: 'red'}

    # Crear un array con los colores correspondientes
    colors_activity = [colors[i] for i in dataset['cluster']]

    # Crear un array de la misma longitud que los datos pero con todos los elementos siendo 1
    # Esto es para que la altura de todas las barras sea la misma
    activity_heights = np.ones(len(dataset))

    # Definir el tamaño de la figura
    plt.figure(figsize=(20, 2))

    # Crear la gráfica
    plt.bar(range(len(dataset)), activity_heights, color=colors_activity, width=1)
    if show_events == True:
        for index, row in events_arrojeria[condition].iterrows():
            event_color = '#00FF00' if row['impact_play'] == 1 else '#93CCEA' if row['impact_play'] == 0  else '#4A0000'
            plt.axvline(x=(row['seconds']-number)/2.5, color=event_color, linestyle='--')

    # Quitar los ticks del eje x para que no se vean los números
    plt.xticks([])
    plt.yticks([])

    # Crear la leyenda
    legend = [mpatches.Patch(color=v, label=k) for k, v in colors.items()]
    plt.legend(handles=legend)

    # Mostrar la gráfica
    plt.show()
```

```python
show_graphic(hr_window_arrojeria_first_time,True,events_arrojeria['seconds']< 2042,0)
```

En la primera parte, en los momentos de máxima intensidad, es cuando más scoring positivo realiza.

```python
hr_window_arrojeria_second_time = hr_window_arrojeria.iloc[round(second_half_time_start/2.5):(round(match_end/2.5))]
```

```python
show_graphic(hr_window_arrojeria_second_time,True,events_arrojeria['seconds']> 2042, 2642)
```

En la segunda parte, Arrojería apenas está en su máxima intensidad y realiza muy pocos eventos.

```python
show_graphic(hr_window_erauskin, False, None, None)
```
