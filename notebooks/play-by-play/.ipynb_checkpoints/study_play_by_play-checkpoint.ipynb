### Event

```python
play_by_play_dataset['Event']
```

```python
play_by_play_dataset['Event'].unique()
```

Etiqueta del evento. Significado:

* FIRST_HALF_START: Inicio de primera parte
* FOUL_RECEIVE: Falta recibida
* POST_OUT: Balón lanzado que acaba al poste y no anota gol
* LOST_BALL: Pérdida de balón
* GOAL: Gol
* 7M_PROVOKE: Falta de 7 metros provocada
* SAVE: Parada
* 7M_PROVOKE_SUSPENSION: Suspensión y falta de 7 metros provocada
* 2_MIN: Exclusión de dos minutos
* FIRST_HALF_END: Final de primera parte
* SECOND_HALF_START: Inicio de segunda parte
* 2MIN_PROVOKE: Provocar una exclusión de dos minutos
* SECOND_HALF_END: Fin de segunda parte

### Home Team

```python
play_by_play_dataset['Home Team']
```

```python
play_by_play_dataset['Home Team'].isnull().any()
```

```python
home_players = []
for index, row in play_by_play_dataset.iterrows():
    event_players = row['Home Team'].split(', ')
    home_players += [player for player in event_players if player not in home_players]
```

```python
home_players
```

Indica las jugadoras del equipo local (Bera Bera) que se encontraban en campo en el evento.

### Visitor Team

```python
play_by_play_dataset['Visitor Team']
```

```python
play_by_play_dataset['Visitor Team'].isnull().any()
```

```python
visitor_players = []
for index, row in play_by_play_dataset.iterrows():
    event_players = row['Visitor Team'].split(', ')
    visitor_players += [player for player in event_players if player not in visitor_players]
```

Indica las jugadoras del equipo visitante (Granollers) que se encontraban en el campo en el evento.
