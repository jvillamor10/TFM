### Se obtiene el valor IsOnCourt

Nombre de la jugadora: Esther Arrojeria

```python
play_by_play_dataset.columns
```

Se debe buscar todos los eventos en los que la jugadora esté en pista, eso se puede saber con la variable Home Team

```python
play_by_play_dataset[play_by_play_dataset[' Home Team'].str.contains('ARROJERIA')]
```

Añadir una columna indicando si está o no en esa jugada

```python
play_by_play_dataset['Is On Court'] = np.where(play_by_play_dataset[' Home Team'].str.contains('ARROJERIA'), 1, 0)
```

```python
play_by_play_dataset
```

Ahora se convertirán los rangos de tiempos a segundos, para que sea más fácil para su tratamiento en otros datasets

```python
df = play_by_play_dataset[[' Start', ' End', ' Event', ' Assist' , ' Offense Player Name', ' Defense Player Name', 'Is On Court']]
df
```

```python
def convert_time(not_formated_time):
    time_values = not_formated_time.split(':')
    return int(time_values[1])*60 + int(time_values[2])
```

```python
convert_time('00:01:24')
```

```python
df[' Start'] = df[' Start'].apply(lambda x: convert_time(x))
```

```python
df[' End'] = df[' End'].apply(lambda x: convert_time(x))
```

```python
df
```

Hay que hacer lo siguiente:
* Añadir el descanso (justo entre los eventos FIRST_HALFTIME_END y SECOND_HALFTIME_START) y además, sumar 15 minutos (duración del descanso) a los registros posteriores a SECOND_HALFTIME_START
* Por cada tiempo muerto y por cada 2_min, añadir 60 registros nuevos

```python
df.loc[85:]
```

Se añade el tiempo de descanso

```python
line = pd.DataFrame({' Start': 1800, ' End': 2400, ' Event': 'REST_TIME','Is On Court': 0}, index=[89])
df = pd.concat([df.iloc[:89], line, df.iloc[89:]]).reset_index(drop=True)
```

```python
df.loc[85:92]
```

```python
df.loc[90:]
```

A partir del registro 90, hay que sumar 900 segundos (tiempo transcurrido de descanso)

```python
df.loc[90:][' Start'] = df.loc[90:][' Start'].apply(lambda x: x + 600)
```

```python
df.loc[90:][' End'] = df.loc[90:][' End'].apply(lambda x: x + 600)
```

```python
df.loc[90:]
```

Se procede a obtener un nuevo dataset, con una columna que equivalga a los segundos y otra columna que equivalga a si la jugadora está en pista o no

Lo primero que hay que hacer, aunque ya se haya sumado el tiempo de descanso, es sumar los tiempos muertos, tanto los explícitos como los de las exclusiones de dos minutos.

```python
df[(df[' Event'] == ' 2_MIN') | (df[' Event'] == 'TIMEOUT')]
```

```python
for i in df[(df[' Event'] == ' 2_MIN') | (df[' Event'] == 'TIMEOUT')].index:
    df.loc[i+1:][' Start'] = df.loc[i+1:][' Start'].apply(lambda x: x + 60)
    df.loc[i:][' End'] = df.loc[i:][' End'].apply(lambda x: x + 60)
```

```python
df
```

```python
df[df[' Event'] == ' SECOND_HALF_START']
```

Ahora, se coge cada registro y se calcula el tiempo que transcurre, y se multiplica ese tiempo por el valor de is on court

```python
is_on_court_list = []
for index, row in df.iterrows():
    diff = row[' End'] - row[' Start']
    if diff == 0 and row[' Event'] not in [' FIRST_HALF_START', ' FIRST_HALF_END', ' SECOND_HALF_START', ' SECOND_HALF_START']:
        diff =1
    list.extend(is_on_court_list,list(str(row['Is On Court']) * diff))
```

```python
len(is_on_court_list)
```

```python
seconds = list(range(1, len(is_on_court_list) +1))
```

Se añade más información para saber los eventos que ocurren

```python
is_on_court_df = pd.DataFrame({}, columns = ['seconds', 'is_on_court'])
is_on_court_df['seconds'] = seconds
is_on_court_df['is_on_court'] = is_on_court_list
is_on_court_df
```

```python
is_on_court_df['event'] = ''
is_on_court_df['assist'] = ''
is_on_court_df['offense_player'] = ''
is_on_court_df['defense_player'] = ''
```

```python
is_on_court_df
```

```python
is_on_court_df.loc[1, 'event']
```

Se añaden los eventos en el segundo que ocurren en el partido

```python
for index, row in df.iterrows():
    value = row[' End'] - 1
    
    if value > 0:
        value +=1
        while is_on_court_df.loc[value, 'event'] != '':
            value += 1
    is_on_court_df.at[value, 'event'] = row[' Event']
    is_on_court_df.at[value, 'assist'] = row[' Assist']
    is_on_court_df.at[value, 'offense_player'] = row[' Offense Player Name']
    is_on_court_df.at[value, 'defense_player'] = row[' Defense Player Name']
```

Se guarda el conjunto de datos

```python
is_on_court_df.to_csv('../../datasets/play-by-play-study/is_player_on_court.csv')
```
