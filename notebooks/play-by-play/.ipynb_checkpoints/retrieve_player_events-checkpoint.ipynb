# Cuaderno que sirve para obtener el impacto de la jugadora en el partido por cada evento

```python
import pandas as pd
```

```python
pd.options.display.max_columns = None
```

```python
PLAY_BY_PLAY_DATASET_FILE_DIR = '../../datasets/play-by-play-study/is_player_on_court.csv'
```

```python
play_by_play_dataset = pd.read_csv(PLAY_BY_PLAY_DATASET_FILE_DIR, index_col=0)
play_by_play_dataset
```

Se van a obtener los eventos de Arrojería

```python
player_name = 'ARROJERIA'
```

Se tiene que tener en cuenta que aparezca como jugadora ofensiva, como jugadora defensiva y como asistente.

```python
player_events = play_by_play_dataset[play_by_play_dataset['assist'].str.contains('ARROJERIA') | play_by_play_dataset['offense_player'].str.contains('ARROJERIA') | play_by_play_dataset['defense_player'].str.contains('ARROJERIA')]
player_events
```

```python
print(f'Hay un total {len(player_events)} para la jugadora {player_name}')
```

hay que ir jugada por jugada y etiquetar si es positiva, neutral o negativa.

```python
player_name = ' 03-E. ARROJERIA'
```

```python
def get_event_impact(event, offense_player, defense_player, assist_player):
    # si provoca una falta, jugada positiva
    if event == ' FOUL_RECEIVE' and offense_player == player_name:
        return 1
    # si comete una falta, jugada neutra
    if event == ' FOUL_RECEIVE' and defense_player == player_name:
        return 0
    # si lanza un balón al poste, jugada negativa
    if event == ' POST_OUT' and offense_player == player_name:
        return -1
    # si recibe un balón al poste, jugada neutra
    if event == ' POST_OUT' and defense_player == player_name:
        return 0
    # si recupera el balón, jugada positiva
    if event == ' LOST_BALL' and defense_player == player_name:
        return 1
    # si pierde el balón, jugada negativa
    if event == ' LOST_BALL' and offense_player == player_name:
        return -1
    # si anota gol, jugada positiva
    if event == ' GOAL' and offense_player == player_name:
        return 1
    # si da una asistencia de gol, jugada positiva
    if event == ' GOAL' and assist_player == player_name:
        return 1
    # si recibe un gol, jugada negativa
    if event == ' GOAL' and defense_player == player_name:
        return -1
    # si provoca un disparo de 7m, jugada positiva
    if event == ' 7M_PROVOKE_SUSPENSION' and offense_player == player_name:
        return 1
    # si comete un disparo de 7m, jugada negativa
    if event == ' 7M_PROVOKE_SUSPENSION' and defense_player == player_name:
        return -1
    # si le paran un disparo a puerta, jugada negativa
    if event == ' SAVE' and offense_player == player_name:
        return -1
    # si para un disparo a puerta, jugada positiva
    if event == ' SAVE' and defense_player == player_name:
        return 1
    # si provoca una exclusión de dos minutos, jugada positiva
    if event == ' 2_MIN' and offense_player == player_name:
        return 1
    # si comete una exclusión de dos minutos, jugada negativa
    if event == ' 2_MIN_PROVOKE' and offense_player == player_name:
        return -1
    # si provoca una exclusión de dos minutos, jugada positiva
    if event == ' 2_MIN_PROVOKE' and offense_player == player_name:
        return 1
    # si comete una exclusión de dos minutos, jugada negativa
    if event == ' 2_MIN_PROVOKE' and offense_player == player_name:
        return -1
```

```python
player_events['impact_play'] = player_events.apply(lambda x: get_event_impact(x['event'], x['offense_player'],x['defense_player'],x['assist']),axis=1)
```
