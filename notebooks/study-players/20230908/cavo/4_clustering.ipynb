# Clustering (HR)

## KMEANS

```python
file = open('../../../../modelos/clustering/hr/kmeans_model.pickle', 'rb')
km = pickle.load(file)
file.close()
```

```python
y_km = km.fit_predict(hr_window)
```

```python
km.labels_
```

```python
hr_window['cluster'] = km.labels_
```

```python
hr_window
```

```python
hr_window.groupby('cluster').describe()
```

```python
mapping = {0:4, 1:2, 2:3, 3:1}
```

```python
hr_window['cluster'] = [mapping[i] for i in hr_window['cluster']]
```

```python
hr_window.groupby('cluster').describe()
```

```python
hr_window.head()
```

```python
hr_window.to_csv(f'../../../../datasets/study-players/{record_date}/clustering/{player_name}_hr_windows_10_50_k_4_kmeans.csv')
```
