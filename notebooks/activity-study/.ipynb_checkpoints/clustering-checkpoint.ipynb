# Clustering

## KMEANS

Primeramente, se va aplicar el algoritmo kmeans

### ActivityLoad2D

Se empezará primero con el activityLoad2D

Se inicializan el parámetro de inicio y de iteración

```python
init = 'random' 
iterations = 10
```

Se procede a calcular la distorsión y el silhouette para cada k-ésimo valor

```python
distortions = []
silhouettes = []

for i in range(2, 11):
    km = KMeans(n_clusters = i, init = init, n_init = iterations)
    labels = km.fit_predict(al2d_window)
    distortions.append(km.inertia_)
    silhouettes.append(metrics.silhouette_score(al2d_window, labels))
```

```python
x = [i for i in range(2,11)]
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('k')
ax1.set_ylabel('sse', color=color)
ax1.plot(x, distortions, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx() 

color = 'tab:blue'
ax2.set_ylabel('silhouette', color=color) 
ax2.plot(x, silhouettes, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()
```

Como se puede ver, el valor con mayor silhouette es k=3, pero, pensando en el problema y viendo que la diferencia entre el silhouette y el sse es muy parecido con K igual a 4, se escoge este último valor.

```python
k = 4
```

```python
print ("Número de clusters", k)
km = KMeans(n_clusters = k, init = init)

y_km = km.fit_predict(al2d_window)
```

## Gaussian Mixture

Se procede a realizar clustering con el algoritmo Gaussian Mixture

```python
gmm = GaussianMixture()
```

Se definen unos cuantos atributos

### ActivityLoad 2D

Se obtienen los mejores parámetros, recorriendo un rango de número de grupos y probando con diferentes covarianzas, escogiendo aquella agrupación con menor BIC.

```python
fig, ax = plt.subplots(figsize=(6, 3.84))

n_components = range(1, 10)
covariance_types = ['spherical', 'tied', 'diag', 'full']

for covariance_type in covariance_types:
    valores_bic = []
    
    for i in n_components:
        modelo = GaussianMixture(n_components=i, covariance_type=covariance_type)
        modelo = modelo.fit(al2d_window)
        valores_bic.append(modelo.bic(al2d_window))
        
    ax.plot(n_components, valores_bic, label=covariance_type)
        
ax.set_title("Valores BIC")
ax.set_xlabel("Número componentes")
ax.legend();
```

El valor que más interesa es el 4, BIC muy bajo y el mismo número de componentes que en HR.

```python
number_components = 4
covariance_type = 'full'
```

Se procede a obtener los clusters

```python
gmm = GaussianMixture(n_components=number_components, covariance_type=covariance_type)
y_gmm = gmm.fit_predict(al2d_window)
```

```python
clusters = cluster.hierarchy.linkage(mat_sim, method = 'complete')
dend = cluster.hierarchy.dendrogram(clusters, labels = al2d_window.index, color_threshold=cut)

f = plt.figure()
plt.show()
```

```python
labels = cluster.hierarchy.fcluster(clusters, cut , criterion = 'distance')
```

```python
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print(f'Número de clusters {n_clusters}')
print("Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(al2d_window, labels))
```
